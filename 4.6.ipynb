```python
from sympy import *

# 题： 解下列方程组，并用 ezplot 函数画出各个方程所对应的直线及交点。Matlab解：
# (a) close all, ezplot('x+3*y=5'), hold on, ezplot('3*x-y=2'), grid on 有解方程组
# (b) close all, ezplot('x+y=2'), hold on, ezplot('3*x+3*y=6'), grid on 不独立方程组
# (a) close all, ezplot('x+y=2'), hold on, ezplot('3*x+3*y=5'), grid on, 不相容方程组

# (a)
x,y = symbols('x y')

p1 = plot(-1/3*x+5/3, show=False)
p2 = plot(3*x-2, show=False)
p1.append(p2[0])
p1.show()

solve([x+3*y-5, 3*x-y-2],(x,y))
```

```python
# (b)

p3 = plot(-1*x+2, show=False)
p4 = plot(-3*x/3+6/3, show=False)
p3.append(p4[0])
p3.show()

solve([x+y-2, 3*x+3*y-6],(x,y))
```

```python
# (c)

p5 = plot(-x+2, show=False)
p6 = plot(-x+5/3, show=False)
p5.append(p6[0])
p5.show()

solve([x+y-2, 3*x+3*y-5],(x,y))
```
